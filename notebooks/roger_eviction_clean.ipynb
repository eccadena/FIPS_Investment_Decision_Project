```python
# Import necessary libraries and dependencies
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
# Use the Pathlib libary to set the path to the CSV
csv_path = Path('../data/rawdata/eviction_data.csv')
df = pd.read_csv(csv_path)
df['GEOID'] = df['GEOID'].astype(str)
df.head()
```

```python
df_data = df[df['parent-location'] == "Harris County, Texas"]
df_data.head()
```

```python
#get columns list to drop unnecesary columns
list(df_data)
```

```python
# Drop extraneous columns
df_data = df_data.drop(columns=[
     'rent-burden',
     'pct-white',
     'pct-af-am',
     'pct-hispanic',
     'name',
     'parent-location',
     'pct-am-ind',
     'pct-asian',
     'pct-nh-pi',
     'pct-multiple',
     'pct-other',
     'renter-occupied-households',
     'eviction-filings',
     'pct-renter-occupied',
     'median-gross-rent',
     'low-flag',
     'eviction-filing-rate',
     'imputed',
     'subbed'
])
df_data = df_data.dropna()
df_data.head()
```

```python
#view columns left
list(df_data)
```

```python
#filter df by Harris County
df_data = df_data.rename(columns={"GEOID":"FIPS"})
df_filtered = df_data[df_data['year'] == 2016]

#filter usign LEN >12 for the FIPS, this avoid duplication of values in data 
df_filtered = df_filtered[df_filtered['FIPS'].str.len()==11]
df_filtered = df_filtered.set_index("FIPS")
df_filtered.head()
```

```python
#groupby to confirm population match 
df_year = df_filtered.groupby('FIPS').sum()
df_year
```

```python
#data frame to csvfile
df_filtered.to_csv(r'../data/cleandata/eviction_data.csv')
```
