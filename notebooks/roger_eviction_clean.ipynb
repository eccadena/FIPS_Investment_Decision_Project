```python
# Import necessary libraries and dependencies
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
# Use the Pathlib libary to set the path to the CSV
csv_path = Path('../data/rawdata/eviction_data.csv')
df = pd.read_csv(csv_path)
df['GEOID'] = df['GEOID'].astype(str)
df.head()
```

```python
#get columns list to drop unnecesary columns
list(df)
```

```python
# Drop extraneous columns
df_data = df.drop(columns=[
     'rent-burden',
     'pct-white',
     'pct-af-am',
     'pct-hispanic',
     'pct-am-ind',
     'pct-asian',
     'pct-nh-pi',
     'pct-multiple',
     'pct-other',
     'low-flag',
     'imputed',
     'subbed'
])
df_data = df_data.dropna()
df_data.head()
```

```python
#view columns left
list(df_data)
```

```python
#filter df by Harris County
df_filtered = df_data.loc[df_data['parent-location'] == "Harris County, Texas"]

#filter usign LEN >12 for the FIPS, this avoid duplication of values in data 
df_filtered = df_filtered[df_filtered['GEOID'].str.len()>11]
df_filtered.head()
```

```python
#groupby to confirm population match 
df_year = df_filtered.groupby('year').sum()
df_year
```

```python
df_filtered.rename(columns={"GEOID":'FIPS'}, inplace=True)
df_filtered.head()
```

```python
#data frame to csvfile
df_filtered.to_csv(r'../data/cleandata/eviction_data.csv', index=False)
```
