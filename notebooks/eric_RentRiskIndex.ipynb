```python
full_dataset = pd.read_csv(Path('../data/cleandata/join_data.csv'))
```

```python
def Rent_Risk_Index(df):
    #retrieve all columns
    #list(df.columns)
    mhi = df['median-household-income']
    PVP = 0
    if mhi > 233000:
        PVP = 1
    elif (mhi <= 233000 and mhi > 118000):
        PVP = .95
    elif (mhi <= 118000 and mhi > 70800):
        PVP = .8
    elif (mhi <= 70800 and mhi > 55600):
        PVP = .6
    elif (mhi <= 55600 and mhi > 43500):
        PVP = .5
    elif (mhi <= 43500 and mhi > 23400):
        PVP = .4
    elif (mhi <= 23400):
        PVP = .2
    #weights associated with attributes. M  - Minority Status
    #                                    Po - Poverty Rate
    #                                    P  - Property Value
    #                                    S  - SVI
    #                                    F  - Food Desert
    #                                    R  - School Rating
    
    M = 0
    Po = 0
    P = 0
    S = 0 
    F = 0
    R = 0
    
    RRI = M*df['Minority Status - Language'] - Po*(df['poverty-rate']/22) + P*PVP - S*df['SVI INDEX'] + R*(df['Rating']) - F*(df['food_desert']*5)
    return RRI
```

```python
full_dataset['RRI'] = full_dataset.apply(Rent_Risk_Index, axis = 1)
```

```python
copy_full_df = full_dataset
for column in copy_full_df.columns.to_list():
    df = copy_full_df
    df[column].fillna((df[column].mean()), inplace = True)
```
