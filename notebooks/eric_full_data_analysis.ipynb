```python
def scatter_plot_vs_poverty(attribute_one):
    df = full_dataset[['poverty-rate', attribute_one]]
#     df.reset_index(inplace=True)
#     df.drop(columns = ['FIPS'], inplace=True)
#     df.set_index(attribute_one, inplace=True)
    
    return df.hvplot('poverty-rate', attribute_one, title = f'Harris County: {attribute_one} vs Poverty Rate, 2016', kind='scatter')

interact(scatter_plot_vs_poverty, attribute_one = full_dataset.drop(columns = ['FIPS','year','Zip']).columns, )
```

```python
def attribute_vs_median_income(attribute_one):
    
    df = full_dataset[[attribute_one, 'median-household-income']]
    df.reset_index(inplace=True)
    #df.drop(columns = ['FIPS'], inplace=True)
    df.set_index(attribute_one, inplace=True)
    
    return df.hvplot(attribute_one,'median-household-income', title = f'{attribute_one} vs Poverty Rate', xlabel = '', ylabel = '', kind='scatter')

interact(attribute_vs_poverty, attribute_one = full_dataset.drop(columns = ['year','Zip', 'FIPS']).columns)
```
