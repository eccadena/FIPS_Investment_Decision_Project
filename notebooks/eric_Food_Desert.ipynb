```python
texas_food_desert = food_desert[food_desert['State'] == 'Texas']
texas_food_desert.head()
```

```python
texas_food_desert_harris = texas_food_desert[texas_food_desert['County'] == 'Harris']
texas_food_desert_harris.head()
```

```python
# XLSX has a sheet with information on all Columns. pd.read_clipboard on this sheet plus an added column on whether to keep or drop this column for our MVP.
tract_info = pd.read_clipboard()
tract_info.head()
```

```python
col_drop = tract_info[tract_info['Keep?'] == 'n']
list_col_drop = col_drop['Field'].to_list()
list_col_drop
```

```python
TX_Food_Desert_Data = texas_food_desert_harris.drop(columns = list_col_drop)
TX_Food_Desert_Data.reset_index(inplace=True)
TX_Food_Desert_Data.rename(columns = {'CensusTract':'FIPS'}, inplace = True) 
TX_Food_Desert_Data.set_index('FIPS', inplace=True)
TX_Food_Desert_Data
```

```python
rural = 0
urban = 0

for entry in TX_Food_Desert_Data['Urban']:
    if entry == 1:
        #print("is Urban")
        urban += 1
    elif entry == 0:
        #print("is Rural")
        rural =+ 1
#     else:
#         print("is Neither Urban or Rural")

print(f'Rural: {rural}')
print(f'Urban: {urban}') #Rural is not necessary since for Harris County... there is only 1 area considered Rural.
```

```python
rural = 0
urban = 0

TX_Food_Desert_Data = TX_Food_Desert_Data[TX_Food_Desert_Data.Urban != 0] #takes up all that Urban data is 1

for entry in TX_Food_Desert_Data['Urban']:
    if entry == 1:
        #print("is Urban")
        urban += 1
    elif entry == 0:
        #print("is Rural")
        rural =+ 1
#     else:
#         print("is Neither Urban or Rural")

print(f'Rural: {rural}')
print(f'Urban: {urban}') #Rural removed
```

```python
TX_Food_Desert_Data.drop(columns = ('Urban'), inplace = True) #because they are all Urban, redundant
```

```python
TX_Food_Desert_Data['food_desert'] = TX_Food_Desert_Data.sum(axis = 1)
```

```python
TX_Food_Desert_Data['food_desert']
```

```python
TX_Food_Desert_Index = TX_Food_Desert_Data.drop(columns = ['LATracts_half','LATracts1', 'LATracts10', 'LATracts20'])
TX_Food_Desert_Index
```

```python
from pathlib import Path

output_path = Path("../data/cleandata/food_desert.csv")

with open(output_path, 'w') as file:
    file.write(TX_Food_Desert_Index.to_csv())
```
