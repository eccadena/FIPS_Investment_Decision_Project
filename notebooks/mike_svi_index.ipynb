```python
import pandas as pd
from pathlib import Path
```

```python
csv_path = Path('../data/rawdata/texas_svi_raw.csv')
svi_df = pd.read_csv(csv_path)
for col in svi_df.columns:
    print(col)
```

```python
svi_clean_df = svi_df[['FIPS','COUNTY','RPL_THEME1','RPL_THEME2','RPL_THEME3','RPL_THEME4','RPL_THEMES']]
svi_clean_df = svi_clean_df.loc[svi_clean_df['COUNTY'] == 'Harris']
svi_clean_df = svi_clean_df.drop(columns='COUNTY')
svi_clean_df.columns=['FIPS','Socioeconomic','Housing - Disability','Minority Status - Language','Housing - Transportation','SVI INDEX']
svi_clean_df.head()
```

```python
svi_clean_df.to_csv(r'../data/cleandata/svi_index.csv', index=False)
```

```python
fips_harris_co = svi_clean_df[['FIPS']]
```

```python
fips_harris_co.index = range(len(fips_harris_co))
```
