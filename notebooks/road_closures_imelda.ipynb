```python
import pandas as pd
from pathlib import Path
import requests
import json
```

```python
#read in road closures data from csv
csv_path = Path('../data/rawdata/Road_Flooding_Imelda.csv')
roads_df = pd.read_csv(csv_path)
roads_df.tail()
```

```python
#Getting FIPS from Lat, Long of road closures
lat_list = roads_df.Lat.to_list()
long_list = roads_df.Long.to_list()
#print(long_list)
```

```python
fips_list_roads = []

for lat,long in zip(lat_list, long_list):
    lat_url = lat
    long_url = long
    url = f"https://geo.fcc.gov/api/census/block/find?latitude={lat_url}&longitude={long_url}&format=json"
    fips_req = requests.get(url)
    fips_res = json.loads(fips_req.text)
    fips_found = fips_res['Block']['FIPS'][:-4]
    fips_list_roads.append(fips_found)
    
```

```python
print(len(fips_list_roads))
```

```python
roads_df['FIPS'] = fips_list_roads
roads_df
```

```python
roads_df.to_csv(r'../data/cleandata/road_flooding_imelda.csv', index=False)
```
